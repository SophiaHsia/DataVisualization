# Some basic polts/charts and how to create them

##**Histogram**

Histogram needs these parts:


*   **The title**: describe the information in this histogram.
*   **X-axis:** intervals shows the scale of values, these intervals are also called bins.
*   **Y-axis:** shows the number of times that the values occured in each interval on X-axis.
*   **The bars:** you can just think the bars the rectangles in the histogram, the width of bars is the interval that is covered, and the height of bars is the times that values occured.



```python
plt.hist(x.percent_senior)
plt.show()
```

## **Bar Plot**

Bar charts is to show how data associated with categorical variables.

```python
plt.bar(x.region, x.crime_rate)
plt.show()
```

```python
plt.barh(x.region, x.crime_rate)
plt.show()
```

## **Pie Chart**


*   X: The wedge size
*   labes: a sequence of strings providing the lables for each wedge.
*   Colors: a sequence of colors that the pie chart will cycle.
* Autopct:string, to lable the wedges with their numeric value. looks like `fmt%pct` (it's just decide the numeric decimal should be)



```python
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_share = [20,25,15,10,20]

# Explode: to show which part should be highlighted
# In this case it should be Firm B
Explode = [0, 0.1, 0, 0, 0]

# about the autopct: we only want 1 decimal place, so it is 1.1f 
plt.pie(market_share, explode = Explode, labels = firms, autopct="%1.1f%%")

plt.show()
```

# **Make Plots Descriptive**

## **1. Adding a Title and lables to your plot**

```python
plt.scatter(x.percent_senior, x.crime_rate)

# Add a title
plt.title('Plot of Crime Rate vs Percent Senior')

# Add lables to the axis
plt.xlabel("Percent Senior")
plt.ylabel("Crime Rate")

plt.show()
```

## **2. Add a legend**

So that you can have multiple lines on same plot, add differ them to each other.

```python
plt.plot(x.work_force, x.income, color='r', label ='Work Force')
plt.plot(x.physicians, x.income, label='Physisians')

# Add a legend
plt.legend()

plt.show()
```

# **Formatting The Plots**

## 1. **Changing Colors**

You can define colors in multiple ways:
1.   `color = '#FFDD44'`
2.   `color = '1.0,0.2,0.3'`: use RGB

*   **`linestyle = 'dashed'`**: Use dashline, also you can use **`--`** to replace `dashed`.

##3. **Adding Markers on your line**

You can use 3 markers to mark certain points on your line.


*   **`marker = o`**: circle marker
*   **`marker = ^`**: triangle marker
*   **`marker = x`**: cross marker



```python
plt.plot(x.work_force, x.income, linestyle='--', marker = 'o', color = '#ADA142')

# You can use `'--ro' to define linestyle, color and marker at the same time`

plt.xlabel("Work Force")
plt.ylabel("Income")

plt.show()
```

# **Multiple Plots**

* We can use `.subplot` to add multiple plots in one figure.
* Also you can use `suptitle` to create subplots titles.
* the `.subplot` function returns a fugure and Axes object or an array of Axes objects.

## **Change your plot size**
If we call subplot function without any parameters, a figure object and one axes object will be returned.

```python
fig, ax = plt.subplots()
print(fig, ax)
```

## **How to create 1 row and 2 columns?**



### Method 1
* Provide the **number of rows, number of columns, and index of the subplot**.
* **index=1**: means the subplot shows first.
* **index=2**: means the subplot shows secondly.
* **`plt.suptitle`**: is to add a common centered to the subplots.

```python
# first subplot, apply formating to it, and add a title
plt.subplot(1,2,1)
plt.plot(x.work_force, x.income, "go")
plt.title("Income vs Work Force")

# second subplot, apply formating to it, and add a title
plt.subplot(1,2,2)
plt.plot(x.hospital_beds, x.income, "r^")
plt.title("Income vs Hospital Beds")

# Add a centered title to all figures
plt.suptitle("Sub Plots")
plt.show()
```

### Method 2

* Use figure and axis to create subplots.
* Then set individual title for each subplot.

```python
fig, (ax1,ax2) = plt.subplots(1,2)

fig.suptitle("Horizontally Stacked Subplots")

ax1.plot(x.work_force, x.income, "go")
ax1.set_title("Income vs Work Force")

ax2.plot(x.hospital_beds, x.income, "r^")
ax2.set_title("Income vs Hospital Beds")

plt.show()
```

## **How to create 2 rows and 1 column?**



```python
plt.subplot(2,1,1)
plt.plot(x.work_force, x.income, "go")

plt.subplot(2,1,2)
plt.plot(x.hospital_beds, x.income, "r^")

plt.suptitle("Sub Plots")
plt.show()
```

## **How to create 2 rows and 2 columns?**

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(6,6))
ax[0,0].plot(x.work_force, x.income, marker = 'o')
ax[0,1].plot(x.work_force, x.income,"bo")
ax[1,0].plot(x.work_force, x.income,"--yo")
ax[1,1].plot(x.work_force, x.income,"--g")

plt.show()
```
